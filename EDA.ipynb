### Para la realizacion del Kpi voy a utilizar el archivo de penetracion y Baf que me indican la penetracion por provincia a los hogares con internet 

```python
# Elimino las  columnas sin datos
df_penetracion2 = df_penetracion.drop(df_penetracion.columns[df_penetracion.isna().all()], axis=1)
```

```python
df_penetracion2.head()
```

```python
df_penetracion2.info()
```

```python
# Cambio el tipo de dato y reemplazo comas por los puntos.
df_penetracion2['Accesos por cada 100 hogares'] = df_penetracion2['Accesos por cada 100 hogares'].apply(lambda x: float(str(x).replace(',', '.')))
```

```python
df_penetracion2.head()
```

```python
df_penetracion2.info()
```

```python
# Concatenenado mis dataframe

df_concatenado = pd.concat([df_penetracion2, df_Baf], axis=1)


df_concatenado = df_concatenado.reset_index(drop=True)
```

```python
df_concatenado.head()
```

```python
# guardo el dataframe en un archivo cvs
df_concatenado.to_csv('concatenado.csv', index=False)
```

### Para la realizacion del segundo Kpi voy a utilizar los archivos de velocidad que me indican las mediciones de velocidad de internet por trimestre

```python
# Guardar el DataFrame concatenado en un archivo CSV

df_internet_velocidad = pd.read_csv('Internet_Accesos-por-velocidad.csv')
df_historico_velocidad = pd.read_csv('historico_velocidad_internet.csv')
df_accesovelocidad_local = pd.read_csv('AccesosaInternetfijoporvelocidaddebajadaylocalidad_2776171699363662004.csv')
df_accesovelocidad_pro = pd.read_csv('AccesosaInternetfijoporvelocidadbajadayprovincia_2791741699447503120.csv')

df_velocidad = pd.concat([df_internet_velocidad, df_historico_velocidad, df_accesovelocidad_local, df_accesovelocidad_pro], ignore_index=True)

df_velocidad.to_csv('velocidad.csv', index=False)
```

```python
df_velocidad.head()
```

```python
# Elimino las  columnas sin datos
df_velocidad = df_velocidad.drop(df_velocidad.columns[df_velocidad.isna().all()], axis=1)
```
